# Get Image From URL

```python
from urllib import request
#f = request.urlretrieve("https://raw.githubusercontent.com/dylanjorgensen/data/master/cat/cat100.jpg", "local-filename.jpg")
f = request.urlretrieve("https://i.ytimg.com/vi/tntOCGkgt98/maxresdefault.jpg", "local-filename.jpg")
f = f[0]
```

# Display

### matplotlib

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')
img = plt.imread(f)
plt.imshow(img)
```
