### Using Requests

```python
import requests

url = 'http://geo.craigslist.org/iso/us/'
headers = {'user-agent': 'Mozilla/5.0 (iPhone; CPU iPhone OS 9_2 like Mac OS X)'}
response = requests.get(url, headers=headers)

# We are testing for the code 200
response.status_code
```

```python
# We don't need the rest of the bs4 package
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.content, 'lxml')

soup.prettify
```

### Using Urllib

```python
#import the library used to query a website
from urllib.request import urlopen

html = urlopen("http://victorianhumour.com/jokedb/?offset=40&paging=40")
print(html)
```

```python
# We don't need the rest of the bs4 package
from bs4 import BeautifulSoup

soup = BeautifulSoup(html, "lxml")

soup.prettify
```

# Downloading First

```python
from six.moves import urllib
import os

# Downloads data
path = "victorian-jokes.txt"
if not os.path.isfile(path):
    urllib.request.urlretrieve("https://raw.githubusercontent.com/dylanjorgensen/datasets/master/victorian/victorian-jokes.txt", path)
```

```python
# Reads text to string
with open(path, 'r') as fp:
    raw_txt = fp.read()
    
raw_txt
```
