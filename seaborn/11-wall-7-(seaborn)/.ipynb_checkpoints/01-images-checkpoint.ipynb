```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

### Make a Directory

```python
# Load the os library
import os

# Create a directory
if not os.path.exists('img_align_celeba'):
    os.mkdir('img_align_celeba')
```

### Retreading Images

```python
# # Load from drive
#img = Image.open("flowers.jpg")
```

```python
# Load from web
```

```python
# Download 10 images from a web directory

# Load the request module
import urllib.request

# Now perform the following 10 times:
for img_i in range(1, 11):

    # create a string using the current loop counter
    f = '000%03d.jpg' % img_i

    # and get the url with that string appended the end
    url = 'https://s3.amazonaws.com/cadl/celeb-align/' + f

    # We'll print this out to the console so we can see how far we've gone
    print(url, end='\r')

    # And now download the url to a location inside our new directory
    urllib.request.urlretrieve(url, os.path.join('img_align_celeba', f))
```

### Batch Dimensions

```python
# List of our 10 images
files
```

```python
# A list of image arrays
imgs = [plt.imread(f_i) for f_i in files]

# shape can be called on a specific list entry
imgs[3].shape
```

```python
plt.imshow(imgs[0])
```

```python
# Combine all of our images to have these 4 dimensions by telling numpy to give us an array of all the images.
data = np.array(imgs)
data.shape
```

###  Mean (A new average image of all faces)

```python
# With axis=0 we average down the pile of photos
mean_img = np.mean(data, axis=0)

# The new shape has the characteristics of a single image
mean_img.shape
```

```python
# matplot lib will only understand the image casted to np.uint8
plt.imshow(mean_img.astype(np.uint8))
```

### Standard Devation (Where changes are likely to be)
- We've just shown where changes are likely to be in our dataset of images.  
- Or put another way, we're showing where and how much variance there is in our previous mean image representation.

```python
std_img = np.std(data, axis=0)

plt.imshow(std_img.astype(np.uint8))
```
